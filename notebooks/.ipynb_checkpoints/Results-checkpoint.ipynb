```python
import pandas as pd
import numpy as np
```

```python
def compute_average(df,col):
    return np.mean(df[col].values)
```

# Subject-Biased Implicit Causality Verbs

### Stimuli with no prompt (ends with '.')

```python
IC1_noprompt = pd.read_csv("../results/summary/IC1-noprompt.csv")
IC1_noprompt
```

```python
print("P(subj_reference |subj=female)", compute_average(IC1_noprompt,'subj_ref_female_subj'))
print("P(obj_reference |subj=female)", compute_average(IC1_noprompt,'obj_ref_female_subj'))
```

```python
print("P(subj_reference |subj=male)", compute_average(IC1_noprompt,'subj_ref_male_subj'))
print("P(obj_reference |subj=male)", compute_average(IC1_noprompt,'obj_ref_male_subj'))
```

### Stimuli with connective: "because"

```python
IC1_connective1 = pd.read_csv("../results/summary/IC1-connective1.csv")
IC1_connective1
```

```python
print("P(subj_reference |subj=female)", compute_average(IC1_connective1,'subj_ref_female_subj'))
print("P(obj_reference |subj=female)", compute_average(IC1_connective1,'obj_ref_female_subj'))
```

```python
print("P(subj_reference |subj=male)", compute_average(IC1_connective1,'subj_ref_male_subj'))
print("P(obj_reference |subj=male)", compute_average(IC1_connective1,'obj_ref_male_subj'))
```

### Stimuli with connective: "and, as a result,"

```python
IC1_connective2 = pd.read_csv("../results/summary/IC1-connective2.csv")
IC1_connective2
```

```python
print("P(subj_reference |subj=female)", compute_average(IC1_connective2,'subj_ref_female_subj'))
print("P(obj_reference |subj=female)", compute_average(IC1_connective2,'obj_ref_female_subj'))
```
