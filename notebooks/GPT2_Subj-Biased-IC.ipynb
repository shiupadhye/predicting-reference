```python
import pandas as pd
import numpy as np
```

```python
def compute_average(df,col):
    return np.mean(df[col].values)
```

## No-prompt

```python
np_average_subj_female_subj_pron = {}
np_average_subj_female_obj_pron = {}
np_average_subj_male_subj_pron = {}
np_average_subj_male_obj_pron = {}
```

### Verb: "aggravated"

```python
IC1_noprompt_1 = pd.read_csv("../results/gpt2/IC1/no-prompt/1.csv")
IC1_noprompt_1.head()
```

```python
# female subject
IC1_noprompt_1f = IC1_noprompt_1[0:12]
IC1_noprompt_1f = IC1_noprompt_1f[["stimulus","She","He"]]
# male subject
IC1_noprompt_1m = IC1_noprompt_1[12:24]
```

#### female subject

```python
IC1_noprompt_1f
```

```python
print("P('subj-pron'|subj=female) = ", compute_average(IC1_noprompt_1f,'She'))
print("P('non-subj-pron'|subj=female) = ", compute_average(IC1_noprompt_1f,'He'))
```

```python
np_average_subj_female_subj_pron['aggravated'] = compute_average(IC1_noprompt_1f,'She')
np_average_subj_female_obj_pron['aggravated'] = compute_average(IC1_noprompt_1f,'He')
```
